```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('adult.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
# заменяем "?" на Nan.
data[data == '?'] = np.nan
data
```

```python
data.info()
# в колонках 'workclass', 'occupation', 'native-country' есть пропуски.
```

```python
# замением пропуски на моды.
for col in ['workclass', 'occupation', 'native-country']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```python
# приводим значения 'income' к 0 и 1.
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['income'] = le.fit_transform(data.income.values)
```

```python
X = data.drop(['income'], axis=1)

y = data['income']
```

```python
# приводим категориальные признаки к числовым.
from sklearn import preprocessing

categorical = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'gender', 'native-country']
for feature in categorical:
        le = preprocessing.LabelEncoder()
        X[feature] = le.fit_transform(X[feature])
X.head()        
```

```python
# Стандартизируем.
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X.head()
```

```python
# применяем логистическую регрессию.
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit( X, y )
predictions = model.predict_proba(X)
```

```python
predictions[:5]
```

```python
model.score(X, y)
```

```python
# применяем метод опорных векторов
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X, y)
```

```python
y_pred = svclassifier.predict(X)
y_pred[:10]
```

```python
svclassifier.score(X,y)
```
