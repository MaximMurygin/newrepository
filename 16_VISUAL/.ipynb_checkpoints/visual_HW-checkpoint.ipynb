```python
df = pd.read_csv('netflix_titles.csv')
# df = df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False).reset_index()
# del df['index']
df
```

```python
df2 = df[((df['release_year']) > 2010)]
df2
```

```python
df2 = df2[['type', 'release_year']]
df_movie = df2[((df2['type']) == 'Movie')]
df_tvshow = df2[((df2['type']) == 'TV Show')]
df_movie
```

```python
df_movie = df_movie.groupby(['release_year']).agg({'type':['count']})
df_movie = df_movie.reset_index()
plt.barh(df_movie['release_year'], df_movie['type','count'])
df_movie
```

```python
df_tvshow = df_tvshow.groupby(['release_year']).agg({'type':['count']})
df_tvshow = df_tvshow.reset_index()

plt.barh(df_tvshow['release_year'], df_tvshow['type','count'])
```
